## Question 1

```python
import pandas as pd
```

```python
students = pd.read_csv('student.csv')
```

```python
students.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
students["Result"] = encoder.fit_transform(students["Result"])
```

```python
students['Result']
```

```python
X = students.drop(["Student", "Result"], axis=1)
y = students["Result"]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X, y)
```

```python
new_student = [[6, 7]]
prediction = KNN.predict(new_student)
```

```python
predicted_label = encoder.inverse_transform(prediction)
print("Prediction (label):", predicted_label[0])
```

## Question 2

```python
weather = pd.read_csv('weather.csv')
```

```python
weather.head()
```

```python
encoders = {}
for column in weather.columns:
    le = LabelEncoder()
    weather[column] = le.fit_transform(weather[column])
    encoders[column] = le
```

```python
weather.head()
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
X = weather.drop(["Play"], axis=1)
y = weather["Play"]
```

```python
NB = CategoricalNB()
NB.fit(X, y)
```

```python
encoded_sample = [[2, 2, 0, 0]]
```

```python
prediction = NB.predict(encoded_sample)
```

```python
decoded_prediction = encoders["Play"].inverse_transform(prediction)
print("Prediction:", decoded_prediction[0])
```
